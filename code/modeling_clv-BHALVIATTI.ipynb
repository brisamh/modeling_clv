Speed up categorical EDA:

```python
cat_cols = clv_df.select_dtypes(include='object').columns
cat_cols = cat_cols.drop('Customer')
print(cat_cols)
```

```python
for col in cat_cols:
    plt.figure(figsize=(8, 5))
    sns.countplot(data=clv_df, x=col, order=clv_df[col].value_counts().index)
    plt.title(f'Distribution of {col}')
    plt.xticks(rotation=45, ha='right')
    plt.tight_layout()
    plt.show()
```

```python
for col in cat_cols:
    avg_clv = clv_df.groupby(col)['Customer Lifetime Value'].mean().sort_values(ascending=False)
    print(f"\nAverage Customer Lifetime Value by {col}:")
    print(avg_clv)
```
